```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.scatter(x, y)
plt.title("Simple Dispersion Graphic")
plt.xlabel("Axis X")
plt.ylabel("Axis y")
plt.show()
```

```python
study_hours = [1,2,3,4,5,6,7,8]
exam_score = [50,55,60,65,70,75,80,85]

# creating a dispersion graphic
plt.scatter(x=study_hours, y=exam_score, alpha=0.6, edgecolors='black', s=50)

# calculating de tendency line
z = np.polyfit(study_hours, exam_score, 1)
p = np.poly1d(z)
plt.plot(study_hours, p(study_hours), "r--")

# titles and labels
plt.title("Relation between hours study and exam scores")
plt.xlabel("Study hours")
plt.ylabel("Exam score")

plt.grid(True)
```
