```python
import matplotlib
import matplotlib.pyplot as plt

business_units = ["Branch A", "Branch B", "Branch C", "Branch D", "Branch E"]
revenue = [200, 240, 150, 400, 220]

# definindo posições das barras no eixo X
positions = range(len(business_units))

# creating the graphic bar
plt.bar(positions, revenue, color="skyblue", edgecolor='black')

# add business units names in the axis X
plt.xticks(positions, business_units)

# title and axis labels
plt.title("Annual revenue by business unit", fontsize=16)
plt.xlabel("Business units")
plt.ylabel("Revenue (in million dollars)")

plt.grid(True, axis='y', linestyle="--", alpha=0.7)
```
